```python
ratings = pd.read_csv('BX-Book-Ratings.csv',sep=';',on_bad_lines='error',encoding='latin-1')
```

```python
ratings.head()
```

```python
ratings.rename(columns={'User-ID':'user_id','Book-Rating':'books_rating'},inplace=True)
```

```python
x =ratings['user_id'].value_counts()>200
```

```python
y=x[x].index
```

```python
ratings = ratings[ratings['user_id'].isin(y)]
```

```python
ratings.head()
```

```python
ratings.shape
```

```python
ratings_with_books = ratings.merge(books,on = 'ISBN')
```

```python
ratings_with_books.head()
```

```python
number_rating =ratings_with_books.groupby('title')['books_rating'].count().reset_index()
```

```python
number_rating.head()
```

```python
number_rating.rename(columns={'books_rating':'no of rating'},inplace=True)
```

```python
final_rating= ratings_with_books.merge(number_rating,on ='title')
```

```python
final_rating
```

```python
final_rating = final_rating[final_rating['no of rating']>=50]
```

```python
final_rating.shape
```

```python
final_rating.drop_duplicates(['user_id','title'],inplace=True)
```

```python
final_rating
```

```python
book_pivot = final_rating.pivot_table(columns='user_id',index ='title',values='books_rating')
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
book_pivot
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(book_pivot)
```

```python
sorted(list(enumerate(similarity_score[0])),key= lambda x:x[1],reverse = True)[1:20]
```

```python

def recommend_book(book_name):
    index = np.where(book_pivot.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key= lambda x:x[1],reverse = True)[1:20]
    data=[]
    for i in similar_items:
        item=[]
        temp_df = books[books['title']== book_pivot.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('title')['title']))
        item.extend(list(temp_df.drop_duplicates('title')['author']))
        item.extend(list(temp_df.drop_duplicates('title')['Image-URL-M']))
        
        data.append(item)
    return data
```

```python
recommend_book('1984')
```
